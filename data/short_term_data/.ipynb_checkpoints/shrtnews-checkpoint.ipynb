```python
import os
import time
import random
import pandas as pd
import requests

# Load short-term stocks from uploaded CSV
df_short_term_stocks = pd.read_csv("./nifty50_stocks.csv")

# Create directories for saving news data
os.makedirs("sentiment_data", exist_ok=True)

# NewsAPI Key (Replace with your actual key)
NEWS_API_KEY = "1dd6845e05bb4e86bc7f87d8a3544ba0"

# Function to introduce short random delays (to prevent rate limits)
def wait():
    delay = random.randint(5, 10)  # Shorter delay (5-10 sec)
    print(f"Sleeping for {delay} seconds to avoid hitting rate limits...")
    time.sleep(delay)

# ---------------- STEP 1: FETCH FINANCIAL NEWS USING NEWS API ---------------- #

def fetch_news_api(stock_name):
    """
    Fetches recent news articles for a given stock using NewsAPI.
    """
    base_url = "https://newsapi.org/v2/everything"
    params = {
        "q": f"{stock_name} stock",
        "apiKey": NEWS_API_KEY,
        "language": "en",
        "sortBy": "publishedAt",
        "pageSize": 5,  # Fetch top 5 news articles per stock
        "domains": "moneycontrol.com,economictimes.indiatimes.com,bloomberg.com"
    }

    news_data = []
    
    try:
        response = requests.get(base_url, params=params, timeout=10)
        response.raise_for_status()  # Raises error if request fails
        data = response.json()

        if "articles" in data:
            for article in data["articles"]:
                news_data.append({
                    "Company Name": stock_name,
                    "News Title": article["title"],
                    "News URL": article["url"],
                    "Published Date": article["publishedAt"],
                    "Source": article["source"]["name"]
                })

        return news_data

    except requests.exceptions.RequestException as e:
        print(f"❌ Error fetching news for {stock_name}: {e}")
        return []

# Loop through selected short-term stocks and fetch news
all_news = []
for stock in df_short_term_stocks["Symbol"]:
    news_results = fetch_news_api(stock)
    if news_results:
        all_news.extend(news_results)
    wait()  # Prevents hitting rate limits

# Save news data to CSV
df_news = pd.DataFrame(all_news)
df_news.to_csv("sentiment_data/short_term_news_data.csv", index=False)
print("\n✅ Short-Term News Data Saved Successfully!")

# ---------------- FINAL OUTPUT ---------------- #
print("\n✅ Short-Term Stock News Collection Completed Successfully!")
```

```python
import os
import requests
import pandas as pd
import nltk
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# 🔹 Download VADER lexicon (Required for Sentiment Analysis)
nltk.download("vader_lexicon")

# 🔹 Initialize VADER Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# 🔹 Load Nifty50 Stock List
df_stocks = pd.read_csv("./nifty50_stocks.csv")  # Make sure the CSV is correctly formatted

# 🔹 API Key (Replace with your own from https://newsapi.org/)
NEWS_API_KEY = "32afb9c04d93456d9845315a4df27eee"

# 🔹 Create Main Folder for Storing Data
MAIN_FOLDER = "nifty50_news_sentiment"
os.makedirs(MAIN_FOLDER, exist_ok=True)

# 🔹 Function to Fetch and Analyze News Data
def fetch_news(stock_name):
    """
    Fetch news articles for a given stock and perform sentiment analysis.
    Saves the data in CSV format and creates a sentiment analysis graph.
    """
    print(f"📩 Fetching news for {stock_name}...")

    url = f"https://newsapi.org/v2/everything?q={stock_name}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    response = requests.get(url)

    if response.status_code == 200:
        articles = response.json().get("articles", [])
        news_data = []

        for article in articles:
            title = article["title"]
            description = article.get("description", "")
            content = article.get("content", "")
            published_at = article["publishedAt"]

            # 🔹 Perform Sentiment Analysis
            sentiment_score = sia.polarity_scores(title + " " + description + " " + content)

            # 🔹 Append Data
            news_data.append({
                "Company Name": stock_name,
                "Date": published_at,
                "Title": title,
                "Description": description,
                "Sentiment Score": sentiment_score["compound"],
                "Sentiment": "Positive" if sentiment_score["compound"] > 0 else "Negative" if sentiment_score["compound"] < 0 else "Neutral",
                "URL": article["url"]
            })

        return pd.DataFrame(news_data)

    else:
        print(f"❌ Error fetching news for {stock_name}: HTTP {response.status_code}")
        return pd.DataFrame()

# 🔹 Process Each Stock in CSV
for stock in df_stocks["Symbol"]:
    try:
        # 🔹 Fetch News & Perform Sentiment Analysis
        df_news = fetch_news(stock)

        if df_news.empty:
            print(f"⚠️ No news found for {stock}. Skipping...\n")
            continue

        # 🔹 Create Folder for Each Stock
        company_folder = os.path.join(MAIN_FOLDER, stock)
        os.makedirs(company_folder, exist_ok=True)

        # 🔹 Save CSV File
        csv_path = os.path.join(company_folder, f"{stock}_news_sentiment.csv")
        df_news.to_csv(csv_path, index=False)
        print(f"✅ Saved news data: {csv_path}")

        # 🔹 Generate Sentiment Graph
        plt.figure(figsize=(10, 5))
        df_news.groupby("Sentiment").size().plot(kind="bar", color=["red", "blue", "green"])
        plt.title(f"Sentiment Analysis of {stock} News Articles")
        plt.xlabel("Sentiment")
        plt.ylabel("Number of Articles")
        plt.xticks(rotation=0)

        # 🔹 Save Graph
        graph_path = os.path.join(company_folder, f"{stock}_sentiment_graph.png")
        plt.savefig(graph_path)
        plt.close()  # Close the plot to prevent memory issues
        print(f"📊 Sentiment graph saved: {graph_path}\n")

    except Exception as e:
        print(f"❌ Error processing {stock}: {e}")

# 🔹 Final Completion Message
print("\n✅ All Nifty50 stock news & sentiment analysis completed successfully!")
```
